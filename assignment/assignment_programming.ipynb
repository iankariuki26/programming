2. Use your function to maximize $f(x) = 100 - 2x^2 + 3x$ for values of $a=-1$, $b=1$, and for values of $N=3$, $N=10$, $N=100$, $N=1000$ and $N=5000$.
3. The true maximizer of this function is $.75$ and the true maximum value is $101.125$. Is that what you got in the previous step? Why not? How does the quality of the maximization depend on $N$? How does your computed answer change as $N$ gets larger?

```python
import numpy as np
import matplotlib.pyplot as plt




#1
def find_maximizer(a, b, N, func, show_plot=False):
    x_values = np.linspace(a, b, N)  # Create the grid of values
    y_values = func(x_values)  # Evaluate func on the grid
    
    if show_plot:  # Plot if show_plot is True
        plt.scatter(x_values, y_values)
        plt.show()

    
    max_value = np.max(y_values)  # Determine the maximum value of func on the grid
    max_indices = np.where(y_values == max_value)  # Find the indices of y_values that equal the maximum
    max_x_values = x_values[max_indices]  # Determine the value of x that maximizes y on the grid
    
    return {'maximum': max_value, 'maximizer': max_x_values}  # Return a dictionary with the maximizer and maximum

#2
def function_to_maximize(x):
    return 100 - (2 * x**2) + (3 * x)
a = -1
b = 1

print(find_maximizer(a, b, 3, function_to_maximize, show_plot=True))
print(find_maximizer(a, b, 10, function_to_maximize, show_plot=True))
print(find_maximizer(a, b, 100, function_to_maximize, show_plot=True))
print(find_maximizer(a, b, 1000, function_to_maximize))
print(find_maximizer(a, b, 5000, function_to_maximize))

#3
"""My answer: It is not the same as my previous answer but it is close. I can see that as the size of N increases, our maximizer value 
approaches the true maximizer value of 101.125. The quality of the maximization incraeses as the size of N increases. MY computed answer
gets closer and closer to the true maximizer value as the quality/N of the maximizer increases"""
```
